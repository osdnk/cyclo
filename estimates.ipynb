```python
import sys
from os.path import dirname
sys.path.append("./lattice-estimator")

from estimator import *

Logging.set_level(Logging.LEVEL0)

# lfpp
kappa = 13 # a = a'
secpar = 2^7
m = 2^20 # also 2^20, 2^18
T = 2^6
B = 2^10 # balanced
q = 2^50 # ifma, chagne to 2^64 for larger T
N = 128
gamma = 2^7 # use biased bin challenges 
base = 1
L = 1
z = 2
e = z

# T = 2^10
# q = 2^59


# T = 2^20
# q = 2^85 # RNS needed


# the expansion of B to be accounted for the SIS break
def sis_break_expansiom(b):
    betahat = b * ((2 * gamma) ** L) + L * 2 * b * ((2 * gamma) ** (L - 1))
    delta = (2 * gamma) ** L
    return float(2 * delta * betahat)

extended_m = N * m * log(2*B, 2*base)
inf_b = (B + 2 * T * gamma)
params = SIS.Parameters(n=N * kappa, m=extended_m, q=q, length_bound=sis_break_expansiom(inf_b) * sqrt(extended_m), norm=2)
costs = SIS.estimate(params) 
costs


# single ring element
rq = log(q, 2) * N / 8 / 1024


# Extension commitment
cost_a = kappa * rq * L

print(float(cost_a))


# Range check
cost_b = z * ((log(N, 2) + log(m * log(2*B, 2*base), 2)) * ((2 * base + 1) * log(q, 2))  / 8 / 1024) + L * rq * z
# 2 sumcheck cost


# Unification of challenges
# Degree 3 as 1 for witness MLE, one for relation 
degree = 3

cost_c = (degree - 1) * e * ((log(N / e, 2) * log(m, 2)) * (log(q, 2)) / 8 / 1024) + z * (L + 1) * rq



print("costs: ", float(cost_a + cost_b + cost_c) , "KB")
```
